```python
#Boucler sur les pages
def next_page():
    global url
    if(soup.find("li", {"class": "next"})!=None):
        print(soup.find("li", {"class": "next"}))
        nextPage = soup.find("li", {"class": "next"}).find("a").get("href")
        if(url.find('index.html')!=-1):
            a = url.find('index.html')
            url = url[:a] + nextPage
            return url
        elif(url.find('page')!=-1):
            a = url.find('page')
            url = url[:a] + nextPage
            return url
    else:
        return
    
```

```python
bookList = []
def get_title_and_price(soup):
    category = soup.title.get_text()
    a = category.find("|")
    category = category[5:a-1]
    articles = soup.find_all('article')
    print(articles[:10])
    for article in articles:
        h3 = article.find_all("h3")
        for i in h3:
            article_title = i.find("a")
            title = article_title.get("title")
        article_price= article.find_all(class_='price_color')
        #print(title)    
        #print(article_price)
        str_cells = str(article_price)
        cleantext = BeautifulSoup(str_cells, "lxml").get_text()
        print(title,": ", cleantext)
        bookList.append((title,cleantext[2:-2],category))
    next_page()
        
for i in range(len(links)):
    url = links[i]
    html = urlopen(url)
    soup = BeautifulSoup(html, 'lxml')
    type(soup)
    
    get_title_and_price(soup)
```
